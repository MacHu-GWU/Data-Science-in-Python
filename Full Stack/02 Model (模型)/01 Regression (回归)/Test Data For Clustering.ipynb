```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
n_samples = 1000
X, y = datasets.make_circles(
    n_samples=n_samples, 
    factor=.5, noise=.05,
)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
n_samples = 1000
X, y = datasets.make_blobs(
    n_samples=n_samples,
    centers=4,
)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
n_samples = 1000
X, y = datasets.make_moons(
    n_samples=n_samples,
    noise=0.05,
)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
n_samples = 1000
random_state=170
X, y = datasets.make_blobs(
    n_samples=n_samples,
    random_state=random_state,
)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X = np.dot(X, transformation)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```
