ROC


1. ROC曲线和AUC值常用语衡量2元分类器的性能.
2. 画出ROC曲线要求分类器输出每个样本属于正类和负类的概率, 否则无法画.




```python
import numpy as np
import pandas as pd
from sklearn import datasets

breast_cancel = datasets.load_breast_cancer()
features = breast_cancel.feature_names
labels = breast_cancel.target_names
X = pd.DataFrame(breast_cancel.data, columns=features)
y = pd.Series(breast_cancel.target)
y = 1 - y # make 1 = Positive, 0 = Negative
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

logreg = LogisticRegression(
    penalty="l1",
    C=1,
)
cross_val_score(
    estimator=logreg,
    X=X,
    y=y,
    scoring="accuracy",
    cv=3,
).mean()
```

下面我们自己来人工画一个ROC曲线

```python
logreg.fit(X, y)
logreg.predict_proba(X)

for
```
